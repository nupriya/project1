# Train test split
TO make the x_train,y_train split wwe are going to split the data

```python
happiness_t=pd.DataFrame(hf['Happiness Score'])
happiness_t
```

```python
y=happiness_t
y
```

```python
happiness_data=pd.DataFrame(hf)
```

```python
x=happiness_data
x
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=53)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Now we use diffrent models for prediction.

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
pred=lm.predict(x_test)
print('Predicted result price:',pred)
print('actual price',y_test)
print('error:')
print(mean_squared_error(pred,y_test))
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
from sklearn.svm import SVR

kernellist=['linear','poly','rbf']
for i in kernellist:
    sv=SVR(kernel=i)
    sv.fit(x_train,y_train)
    print(sv.score(x_train,y_train))
```

```python
import sklearn
from sklearn.svm import SVC
from sklearn.metrics import explained_variance_score,r2_score,mean_tweedie_deviance
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

import warnings
warnings.filterwarnings('ignore')

model=[Ridge(alpha=.4),LinearRegression(), Lasso(), DecisionTreeRegressor(max_depth=15,min_samples_leaf=100),RandomForestRegressor(n_estimators=500,max_depth=6,min_samples_leaf=50,n_jobs=4)]


for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print(' Score = \n',m,'is :')
    print(explained_variance_score(y_test,predm))
    print(r2_score(y_test,predm))
    print(mean_tweedie_deviance(y_test,predm))
    print('\n')
```
